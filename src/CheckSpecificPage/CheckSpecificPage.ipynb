# Check a Specifict web side of Fast Check


```python
import requests
from bs4 import BeautifulSoup
import json
from colorama import Fore, Style
```

# Check one web site


```python
link = "https://www.fastcheck.cl/2022/07/13/la-propuesta-de-nueva-constitucion-le-permitira-a-evo-morales-establecer-a-traves-de-la-plurinacionalidad-con-autonomia-y-autogobierno-la-independencia-o-anexion-de-algun-territorio-declarado-de/"
response = requests.get(link)
soup = BeautifulSoup(response.content, "html.parser")

title = soup.find("h1", class_="single-post-title entry-title").text

veracity = title[title.find("#") + 1:]
author = soup.find("ul", class_="meta ospm-modern clr").find("a").text
text = soup.find("div", class_="entry-content clr").text
links = [a["href"] for a in soup.find("div", class_="container clr").find_all("a", href=True)]
images = soup.find("div", class_="entry-content clr").find_all("img", class_="size-full wp-image-")

print("Revisión sin errores")
print(author)
print(veracity)
print(text)
print(links)
print(images)

json_data = {
    "title": title,
    "author": author,
    "veracity": veracity,
    "text": text,
    "links": links,
    "images": images
}

with open("page_checked.json", 'w', encoding='utf-8') as file:
    json.dump(json_data, file, ensure_ascii=False, indent=4)    
```

# Check a json file (Fact Check)


```python
json_name = "2nd_fast_check_pages"
doc_path = f"../fast_check_data/{json_name}.json"
json_data = {}
pages = []

# The file in "doc_path" must be a json file with the following structure:
# {
#    "pages": [
#       {
#           "title": "Title of the page",
#           "link": "Link of the page",
#           "author": "Author of the page",
#           "veracity": "Veracity of the page",
#           "text": "Text of the page",
#           "links": ["Link 1", "Link 2", ...],
#           "images": ["Image 1", "Image 2", ...]
#      },
#     ...
#   ]
# }
# The "veracity" field must be empty or "No se encontró la veracidad" if the page doesn't have a veracity 

with open(doc_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)  
    pages = json_data["pages"]

checked_json_data = json_data

for index, page in enumerate(pages):
    print(f"Revisando página {index + 1}")
    if page["veracity"] == "No se encontró la veracidad" or page["veracity"] =="":
        print(Fore.BLUE + "Página sin veracidad encontrada" + Style.RESET_ALL)
        print(page["title"])
        print(page["link"])

        response = requests.get(page["link"])
        soup = BeautifulSoup(response.content, "html.parser")
        title = soup.find("h1", class_="single-post-title entry-title").text
        if title.find("#") == -1:
            print(Fore.BLUE + "La página no contienen una veracidad, se procede a marcar con 'Remover' " + Style.RESET_ALL)
            checked_json_data["pages"][index]["veracity"] = "Remover"    

        else:
            veracity = title[title.find("#") + 1:]  
            author = soup.find("ul", class_="meta ospm-modern clr").find("a").text
            text = soup.find("div", class_="entry-content clr").text
            links = [a["href"] for a in soup.find("div", class_="container clr").find_all("a", href=True)]
            images = soup.find("div", class_="entry-content clr").find_all("img", class_="size-full wp-image-")

            checked_json_data["pages"][index]["author"] = author
            checked_json_data["pages"][index]["veracity"] = veracity
            checked_json_data["pages"][index]["text"] = text
            checked_json_data["pages"][index]["links"] = links
            checked_json_data["pages"][index]["images"] = images

            print(f"Veracidad: {veracity}")
        print()

with open(doc_path, 'w', encoding='utf-8') as file:
    json.dump(checked_json_data, file, ensure_ascii=False, indent=4)
```
