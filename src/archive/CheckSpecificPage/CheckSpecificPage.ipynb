# Check one web site


```python
import requests
from bs4 import BeautifulSoup
import json
from colorama import Fore, Style
```

```python
link = "https://www.fastcheck.cl/2022/05/17/video-maduro-celebra-y-manda-un-saludo-a-la-convencion-por-la-entrega-del-borrador-pluri-constituyente-enganoso/"
response = requests.get(link)
soup = BeautifulSoup(response.content, "html.parser")

title = soup.find("h1", class_="elementor-heading-title elementor-size-default").text
veracity = title[title.find("#") + 1:]
author = soup.find("div", class_="elementor-element elementor-element-1786e60 elementor-align-left elementor-widget elementor-widget-post-info").find("ul", class_="elementor-inline-items elementor-icon-list-items elementor-post-info").find("li").find("span", class_="elementor-icon-list-text elementor-post-info__item elementor-post-info__item--type-author").text
author = author.replace("\n", "").replace("\t", "").split(":")[1]
text = soup.find("div", class_="elementor-element elementor-element-fb8d216 elementor-widget elementor-widget-theme-post-content").text
links = [a["href"] for a in soup.find("div", class_="container clr").find_all("a", href=True)]
images = soup.find("div", class_="elementor-element elementor-element-fb8d216 elementor-widget elementor-widget-theme-post-content").find_all("img")
images = [figure["src"] for figure in images if figure["src"].split("/")[-1] != "Suscribete-2021-1024x536.png" and figure["src"].split("/")[-1] != "No-copiar-2.png"]
if len(images) > 0:
    del images[0]

print("Revisión sin errores")
print(author)
print(veracity)
print(links)
print(images)

json_data = {
    "title": title,
    "author": author,
    "veracity": veracity,
    "text": text,
    "links": links,
    "images": images
}

with open("page_checked.json", 'w', encoding='utf-8') as file:
    json.dump(json_data, file, ensure_ascii=False, indent=4)    
```

# Check Fast Check's json file


```python
import requests
from bs4 import BeautifulSoup
import json
from colorama import Fore, Style

json_name = "2nd_fast_check_pages"
doc_path = f"../fast_check_data/{json_name}.json"
doc_name = "../fast_check_data/test_2nd.json"
json_data = {}
pages = []

# The file in "doc_path" must be a json file with the following structure:
# {
#    "pages": [
#       {
#           "title": "Title of the page",
#           "author": "Author of the page",
#           "veracity": "Veracity of the page",
#           "text": "Text of the page",
#           "links": ["Link 1", "Link 2", ...],
#           "images": ["Image 1", "Image 2", ...]
#      },
#     ...
#   ]
# }
# The "veracity" field must be empty or "No se encontró la veracidad" if the page doesn't have a veracity 

with open(doc_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)  
    pages = json_data["pages"]

checked_json_data = json_data

for index, page in enumerate(pages):
    print(f"Revisando página {index + 1}")

    try:
        page["link"]
    except KeyError:
        print(Fore.RED + "La página no contiene un link, se procede a marcar con 'Remover' " + Style.RESET_ALL)
        checked_json_data["pages"][index]["veracity"] = "Remover"
        continue

    response = requests.get(page["link"])
    print(page["link"])

    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.find("h1", class_="elementor-heading-title elementor-size-default").text
    if title.find("#") == -1:
        print(Fore.BLUE + "La página no contienen una veracidad, se procede a marcar con 'Remover' " + Style.RESET_ALL)
        checked_json_data["pages"][index]["veracity"] = "Remover"    

    else:
        veracity = title[title.find("#") + 1:]
        author = soup.find("div", class_="elementor-element elementor-element-1786e60 elementor-align-left elementor-widget elementor-widget-post-info").find("ul", class_="elementor-inline-items elementor-icon-list-items elementor-post-info").find("li").find("span", class_="elementor-icon-list-text elementor-post-info__item elementor-post-info__item--type-author").text
        author = author.replace("\n", "").replace("\t", "").split(":")[1]
        text = soup.find("div", class_="elementor-element elementor-element-fb8d216 elementor-widget elementor-widget-theme-post-content").text
        links = [a["href"] for a in soup.find("div", class_="container clr").find_all("a", href=True)]
        images = soup.find("div", class_="elementor-element elementor-element-fb8d216 elementor-widget elementor-widget-theme-post-content").find_all("img")
        images = [figure["src"] for figure in images if figure["src"].split("/")[-1] != "Suscribete-2021-1024x536.png" and figure["src"].split("/")[-1] != "No-copiar-2.png"]
        if len(images) > 0:
            del images[0]

        checked_json_data["pages"][index]["author"] = author
        checked_json_data["pages"][index]["veracity"] = veracity
        checked_json_data["pages"][index]["text"] = text
        checked_json_data["pages"][index]["links"] = links
        checked_json_data["pages"][index]["images"] = images

        print(f"Veracidad: {veracity}")
    print()

with open(doc_name, 'w', encoding='utf-8') as file:
    json.dump(checked_json_data, file, ensure_ascii=False, indent=4)
```
